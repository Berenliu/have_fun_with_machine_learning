```python
'''
According to https://blog.learningtree.com/assumptions-ruin-k-means-clusters/ 
and http://varianceexplained.org/r/kmeans-free-lunch/,
k-means has 2 assumptions:
1. Clusters are spatially grouped—or “spherical”
2. Clusters are of a similar size

For the first assumption, one may consider every cluster is almost simply connect, i.e, the inner area of every cluster 
should contain its elements;
For the second assumption, it is not always a problem. See the following examples, when we choose the cluster 2 with 100 
or 50 observations, k-means works well, but when we reduce the number of observations of cluster 2 to 30, the bad thing comes! 
One interesting thing is that it does not always works badly, the result depends on the iniatialization. However, for the most 
of cases, it works badly! And if we reduce the number to 20, it almost always works badly. Theoretically, k-means can always find
the right cluster structure if we know the correct iniatialization. In practice, this is almost impossible. It seems that the less
observations we have for cluster 2, the more likely k-means will work badly. Also, the last try shows that if the distance between
the two clusters are big enough, k-means will also work well. 
'''     
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plf
import seaborn as sns
from ggplot import *
%matplotlib inline
```

```python
dat1 = 0.01*np.random.randn(1000, 2) + np.array([0.02, 0.02])
dat2 = 0.01*np.random.randn(100, 2) - np.array([0.02, 0.02])
df1 = pd.DataFrame(dat1, columns=list("xy"))
df1["choice"] = "1"
df2 = pd.DataFrame(dat2, columns=list("xy"))
df2["choice"] = "2"
df = pd.concat([df1, df2])
print(df.shape)
kmeans = KMeans(n_clusters=2, max_iter=300, init="random")
cluster_info = kmeans.fit_predict(df[["x", "y"]].values)
df["kmeans_cluster"] = [str(i) for i in cluster_info]
```

```python
ggplot(df, aes(x="x", y="y", color="choice")) + geom_point()
```

```python
ggplot(df, aes(x="x", y="y", color="kmeans_cluster")) + geom_point()
```
